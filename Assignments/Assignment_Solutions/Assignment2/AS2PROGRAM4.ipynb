```python
# PROGRAM4
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('cereal.csv')
d=pd.DataFrame(data)
d
```

```python
mfr=d['mfr']
mfr
```

```python
rt=d['rating']
rt
```

```python
plt.figure(figsize=(10,6))
plt.barh(mfr,rt,color='b')
plt.xlabel('RATING')
plt.ylabel('MANUFACTURERS')
plt.xticks(np.arange(0,105,5))
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=d,x="rating",y="mfr")
plt.xticks(np.arange(0,105,5))
plt.show()
```
